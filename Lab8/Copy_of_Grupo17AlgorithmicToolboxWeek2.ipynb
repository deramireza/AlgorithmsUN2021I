## Last Digit of a Large Fibonacci Number

```python
%%writefile LDLFN.cpp

#include <iostream>

using namespace std;

int grr(int n) {
    if (n <= 1)
        return n;
    int uno = 0;
    int dos  = 1;
    for (int i = 0; i < n - 1; ++i) {
        int temporal = uno;
        uno = dos;
        dos = temporal + dos;
        if(dos>9){
        	dos=dos%10;
        }
    }
    return dos;
}

int main() {
    int n,ultimodigito;
    cin >> n;
    cout << grr(n)<< '\n';
    }
```

##Fibonacci Number 

```python
%%writefile FN.cpp

#include <iostream>
#include <cassert>
#include<math.h>

long long fibonacci_fast(int n) {
    
    long long F[n+1];
    F[0] = 0LL;
    F[1] = 1LL;
    for(int i =2; i<n+2; i++){
     F[i] = (long long)(F[i-1]+F[i-2]);
    } 

    return F[n];
}

int main() {
    int n = 0;
    std::cin >> n;

    std::cout << fibonacci_fast(n) << '\n';
    return 0;
}
```

##Greatest Common Divisor

```python
%%writefile GCD.cpp
#include <iostream>

int gcd(int a, int b) {
  if(b==0)
   return a;
  else
   return gcd(b,a%b);
}

int main() {
  int a, b;
  std::cin >> a >> b;
  std::cout << gcd(a, b) << std::endl;
  return 0;
}
```

##Fibonacci Number Again

```python
%%writefile FNA.cpp
#include <iostream>

using namespace std;

long long int P(long long int m) {
	long long int x = 0, y = 1, z = x + y; 
	for (int i = 0; i < m*m; i++) {
		z = (x + y) % m; 
		x = y; 
		y = z; 
		if (x == 0 && y == 1)
			return i + 1; 
	}
}
int Solucionar(long long int n, long long int m) {
	long long remain = n % P(m);

	long long first = 0;
	long long second = 1;

	long long res = remain;

	for (int i = 1; i < remain; i++) {
		res = (first + second) % m;
		first = second;
		second = res;
	}

	return res % m;
}
int main() {
	long long int n, m; 
	cin >> n >> m; 
	cout << Solucionar(n, m) << endl;
}
```
